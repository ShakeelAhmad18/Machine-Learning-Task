# Task-1
### Let's consider a scenario where we want to determine the linear relationship between the 
### numbers of hours a student studies and the percentage of marks that student scores in an exam. We 
### want to find out that given the number of hours a student prepares for a test, about how high of a 
### score can the student achieve?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
full_data=pd.read_csv('student_scores.csv')
full_data
```

```python
full_data.isnull()
```

```python
full_data.dropna()
```

```python
x=full_data.drop('Scores',axis=1)
y=full_data['Scores']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
x_test
```

```python
lr.score(x_test,y_test)
```

```python
lr.predict([[12]])
```

```python
plt.figure(figsize=(5,6))
plt.scatter(x,y,marker='o')
plt.plot(x,lr.predict(x),label=('Best Fit Line'),color='r')
plt.legend()
plt.show()
```

```python
diff=y_test-y_pred
diff
```

```python
MSE=mean_squared_error(y_test,y_pred)
MSE
```

```python
rmse=np.sqrt(MSE)
rmse
```

```python
sns.distplot(diff,kde=True)
```

#  Task-2

###  In this scenario we will use multiple linear regression to predict the gas consumptions (in 
### millions of gallons) in 48 US states based upon gas taxes (in cents), per capita income (dollars), 
### paved highways (in miles) and the proportion of population that has a driver’s license. Develop a 
### regression model for this task.

```python
petrol=pd.read_csv('petrol_consumption.csv')
petrol.tail()
```

```python
petrol.head()
```

```python
petrol.dropna()
```

```python
petrol.isnull()
```

```python
x=petrol.drop('Petrol_Consumption',axis=1)
y=petrol['Petrol_Consumption']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

```python
x_train
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_predi=model.predict(x_test)
y_predi
```

```python
y_test
```

```python
model.score(x_test,y_test)
```

```python
result=y_test-y_predi
```

```python
result
```

```python
sns.scatterplot(x=y_test,y=y_predi,color='blue',label='Actual Data Points')
plt.plot([min(y_test),max(y_test)],[min(y_test),max(y_test)],color='r',label='Ideal Line')
plt.legend()
plt.show()
```

```python
sns.distplot(result,kde=True)
```

```python
mse=mean_squared_error(y_test,y_predi)
```

```python
mse
```

```python
rmse=np.sqrt(mse)
```
